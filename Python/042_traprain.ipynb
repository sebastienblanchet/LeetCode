042 Trapping Rain

Given n non-negative integers representing an elevation map
where the width of each bar is 1, compute how much water it is 
able to trap after raining.



The above elevation map is represented by array 
[0,1,0,2,1,0,1,3,2,1,2,1]. In this case,
6 units of rain water (blue section) are being trapped. 



Example:

Input: [0,1,0,2,1,0,1,3,2,1,2,1]
Output: 6
    
Topics:
Two pointers, stack

```python
class Solution(object):
    def trap(self, height):
        """
        :type height: List[int]
        :rtype: int
        """
        n = len(height)
        # base case, if there are less than 3 drops of water
        # none can be trapped
        if n <=2: return 0
        # init array of rain drops at each level
        ans = [0]*max(height)
        
        # init left and right pointers
        l = 0
        r = n -1
        
        # iterate through pointers
        while l < r:            
            # move pointers as needed
            print("Left {} and Right {}" .format(l, r))
            
            # get smallest value
            minval = min(height[r], height[l])
            
            for i in range(l, r):
                if height[i] < minval:
                    ans[minval] +=1
            print(ans)
                    
            # move pointers
            # need to go to next greates
            # if right is greater, increment left
            lval = height[l]            
            rval = height[r]
            
            if rval > lval:
                while True:
                    l +=1
                    if lval <= height[l]: break
            # otherwise decrement right until 
            else:
                while True:
                    r -=1
                    if rval <= height[r]: break
        print(ans)
        return sum(ans)
    
height = [0,1,0,2,1,0,1,3,2,1,2,1]
Solution().trap(height)
```
