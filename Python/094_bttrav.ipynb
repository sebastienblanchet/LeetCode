094 Binary Tree Inorder Traversal
https://leetcode.com/problems/binary-tree-inorder-traversal/description/

Given a binary tree, return the inorder traversal of its nodes' values.

Example:

Input: [1,null,2,3]
   1
    \
     2
    /
   3

Output: [1,3,2]
Follow up: Recursive solution is trivial, could you do it iteratively?

Hash table, stack, tree

```python
# Definition for a binary tree node.
from __future__ import print_function

class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        
# recall stack
class Stack:
     def __init__(self):
         self.items = []

     def isEmpty(self):
         return self.items == []

     def push(self, item):
         self.items.append(item)

     def pop(self):
         return self.items.pop()

     def peek(self):
         return self.items[len(self.items)-1]

     def size(self):
         return len(self.items)

class Solution(object):
    
    lst = []
    
    '''
    Recursive solution
    TIME = O(n) --> recursion is T(n) = 2T(n/2)+1
    SPACE = O(n) WORST
    '''
    def inorderTraversal(self, root):
        """
        :type root: TreeNode
        :rtype: List[int]
        """
        
        if root:
            # recur on
            self.inorderTraversal(root.left)
            
            print(root.val)
            self.lst.append(root.val)
            print(self.lst)
            
            # recur on right
            self.inorderTraversal(root.right)
            
        return self.lst
    
    '''
    Iteratively using a stack
    '''

    def inorderTrav(self, root):
        temp = []
        ans = []
        node = root
        cnt = 1
        
        while node or temp:
            while node:
                temp.append(node)
                node = node.left
            node = temp.pop()
            ans.append(node.val)
            node = node.right
            cnt +=1
        return ans
        
    
root = TreeNode(1)
nodea = TreeNode(2)
nodeb = TreeNode(3)

root.right = nodea
nodea.left = nodeb

Solution().inorderTrav(root)
```
